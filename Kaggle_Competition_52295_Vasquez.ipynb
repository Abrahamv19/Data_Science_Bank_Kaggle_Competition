```python
#Definicion de funcion para determinar los nulos en dataset
def nulos(dfTrain):
  cant = len(dfTrain)
  for col in dfTrain.columns:
    print(col, ':', round(dfTrain[col].isnull().sum() * 100 / cant,2), '%')

nulos(dfTrain)
```

```python
dfTrain.isnull().sum()
```

```python
sum(dfTrain.isna().sum())
# no hay nulos
```

```python
#Eliminacion de NANN o Nulos
dfTrain = dfTrain.dropna()
dfTrain
```

```python
# Obtiene los valores únicos de la columna "Geography"
valores_unicos = dfTrain["Geography"].unique()

# Imprime los valores únicos
print(valores_unicos)
```

```python
# Obtiene los valores únicos de la columna "Gender"
valores_unicos2 = dfTrain["Gender"].unique()

# Imprime los valores únicos
print(valores_unicos2)
```

```python
dfTrainCoded =  dfTrain
pais = {'France':1, 'Spain':2, 'Germany':3}
dfTrainCoded["Geography"] = dfTrainCoded["Geography"].map(pais)
```

```python
genero = {'Male':1, 'Female':2}
dfTrainCoded["Gender"] = dfTrainCoded["Gender"].map(genero)
```

```python
#Incluir en el data set solo los numericos
dfTrainNum = dfTrainCoded.select_dtypes(include=[np.number])
dfTrainNum.head(3)
```

```python
#Analisis de matriz de correlacion
dfTrainNum.corr().T
```

```python
dfTrainNum = dfTrainNum.dropna()
```

```python
dataTrain = dfTrainNum.values
```

```python
# separar en x y en y
x, y = dataTrain[:, :-1], dataTrain[:, -1]
print(x.shape, y.shape)
```

```python
# separamos en entrenamiento/test en razon de 70/30
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
#Crear el modelo de regresion lineal
randomForest = RandomForestClassifier(random_state=2)
#Ajustar el modelo con X_train y y_train
randomForest.fit(x_train, y_train)
#Predecir con X_test
y_pred = randomForest.predict(x_test)
```

```python
y_pred.size
```

```python
y_pred
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
# Todas las metricas
print(metrics.classification_report(y_test, y_pred))
```
